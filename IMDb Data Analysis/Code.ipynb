```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
#import the dataset
imdb_data = pd.read_csv("IMDB.csv")
```

```python
# Display basic information about the dataset
print(imdb_data.info())
```

```python
# Display summary statistics of numeric columns
print(imdb_data.describe())
```

```python
# Plot histogram of movie scores
plt.figure(figsize=(10, 6))
sns.histplot(imdb_data['score'], bins=20, kde=True)
plt.title('Distribution of Movie Scores')
plt.xlabel('Score')
plt.ylabel('Frequency')
plt.show()
```

```python
# Plot bar chart of top 10 movie genres
plt.figure(figsize=(12, 8))
top_genres = imdb_data['genre'].value_counts().head(10)
sns.barplot(x=top_genres.values, y=top_genres.index, palette='viridis')
plt.title('Top 10 Movie Genres')
plt.xlabel('Number of Movies')
plt.ylabel('Genre')
plt.show()
```

```python
# Plot scatter plot of movie revenue vs. budget
plt.figure(figsize=(10, 6))
sns.scatterplot(x='budget_x', y='revenue', data=imdb_data)
plt.title('Movie Revenue vs. Budget')
plt.xlabel('Budget')
plt.ylabel('Revenue')
plt.show()
```
